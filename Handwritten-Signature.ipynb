# Read images

```python
x_train=[]
x_test=[]
folders_name=["personA","personB","personC","personD","personE"]
for name in folders_name:
    for img in glob.glob("data/{}/Train/*.png".format(name)):
        n=cv2.imread(img)
        n= cv2.resize(n, (300,300))
        x_train.append(n)
    for img in glob.glob("data/{}/Test/*.png".format(name)):
        n=cv2.imread(img)
        n= cv2.resize(n, (300,300))
        x_test.append(n)
```

```python
x_train[30].shape
#len(cv_img)
```

# create labels

```python
y_train=np.arange(200)
y_test=np.arange(40)
```

```python
def create_labels(step_size,start,end):
    temp=np.arange(end)
    for idx, i in enumerate(range(start, end, step_size)):
        temp[i:i+step_size] = idx
    return temp
```

```python
y_train=create_labels(40,0,200)
y_train
y_test=create_labels(8,0,40)
y_test
```

```python
x_train
```

```python
x_train=np.array(x_train)
x_test=np.array(x_test)
x_train.shape
```

```python
x_test=x_test/255
x_train=x_train/255
```

```python
x_train
```

# Train Model

```python
cnn=models.Sequential([
    layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(300,300,3)),
    layers.MaxPool2D((2,2)),
    
    layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
    layers.MaxPool2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(5,activation='softmax')
])

cnn.compile(
optimizer='adam',
loss='sparse_categorical_crossentropy',
metrics=['accuracy']
)

cnn.fit(x_train,y_train,epochs=5)
```

```python
cnn.evaluate(x_test,y_test)
```

# Test Model

```python
result=cnn.predict(x_test)
```

```python
final_result=[]
for i in result:
    final_result.append(np.argmax(i))
```
